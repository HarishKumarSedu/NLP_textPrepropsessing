```python
with open('MovieData.json')as file:
    movie_data = json.load(file) 
```

```python
data = {}
for i,record in enumerate(movie_data):
    data[i]=record
data
```

```python
import pandas as pd 

df = pd.DataFrame.from_dict(
   data=data, orient = 'index',columns=['name', 'description', 'genre']
)
```

```python
df
```

```python
# Convert into the Lower case
df.description = df.description.str.lower()
```

```python
df.description
```

```python
# Remove the HTML tags 
import re 
df.description = df.description.apply( lambda x:re.compile(r'<.*?>').sub('',x) )
df.name = df.name.apply( lambda x:re.compile(r'<.*?>').sub('',x) )
df
```

```python
# Remove the URLS from the text 
df.description = df.description.apply( lambda x:re.compile(r'https?://\S+|www\.\S+http?://\S').sub('',x) )
```

```python
import string
string.punctuation
y=lambda x: x.translate(str.maketrans('', '', string.punctuation))
```

```python

df.description = df.description.apply( lambda x: x.translate(str.maketrans('', '', string.punctuation)) )
```

```python
df.name = df.name.apply( lambda x: x.translate(str.maketrans('', '', string.punctuation)) )
```

```python
df
```

```python
# remove emojis from the description 
df.description = df.description.apply(
    
    lambda x: re.compile("["
                           u"\U0001F600-\U0001F64F" # emoticons
                           u"\U0001F300-\U0001F5FF" # symbols & pictographs
                           u"\U0001F680-\U0001F6FF" # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF" # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE).sub('',x)
)
```
